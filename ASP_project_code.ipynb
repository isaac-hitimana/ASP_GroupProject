```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
arrival_rate = range(15,51)
print(arrival_rate)
```

```python
service_rate = 30
```

```python
system_utilization = []
for i in arrival_rate:
    system_utl = i/service_rate
    system_utilization.append(system_utl)

print(system_utilization)  
```

```python
plt.figure(figsize = (9,8))
plt.bar(arrival_rate, system_utilization, color = 'g')
plt.axvline(30, color = 'r', label = ('Maximum cpacity'))
plt.axhline(1.0, color = 'r')

plt.legend()
plt.title(" arrival rates vs system utilisation $\%$")
plt.xlabel("Arrival rates")
plt.ylabel("System Utilisation in Percentages")
plt.show()
```

```python
roots = [30,1, 1.2]

plt.figure(figsize = (11,10))

plt.style.use('ggplot') # Using ggplo2 library to plot an oustanding plot 

plt.annotate(" Maximum Capacity of the system", xytext =(26, 1.4),#  Text coordinate 
             xy = (30,1.0), arrowprops={'facecolor':'black'}) # Locating the break even point 

plt.annotate(" Initial investiment",
             
             xytext =(20, 2), xy = (0,2), arrowprops={'facecolor':'black'}) # Initial investiment

plt.plot(arrival_rate, system_utilization, 'g-',
         markevery = roots, label='Utilisation rates')


# Create legend.4
plt.legend(loc='upper left')
plt.xlabel('Arrival Rates')
plt.ylabel('Utilisation in $\%$')
plt.title(' Arrivaal rates vs System Utilisation')

plt.savefig('utilisation.png')

plt.show()
```

```python
average_customers_sys = []
num_arrival_rate = []
for i in arrival_rate:
    if i < service_rate:
        average_customer = i/(service_rate - i)
        average_customers_sys.append(average_customer)
        num_arrival_rate.append(i)

    
for j in arrival_rate:
    if j > service_rate:
        average_customer = j/(service_rate - j)
        average_customers_sys.append(average_customer)
        num_arrival_rate.append(j)

print(average_customers_sys)
print(num_arrival_rate)
```

```python
average_time_sys = []
for i in num_arrival_rate:
    time_sys = 1 /(service_rate - i)*60
    average_time_sys.append(time_sys)

print(average_time_sys)
```

```python
prob_sys_empty = []

for i in system_utilization:
    p_sys_empty = 1 - i 
    prob_sys_empty.append(p_sys_empty)
    
print(len(prob_sys_empty))
print(len(system_utilization))
```

```python
prob_sys_empty
df = pd.DataFrame()
df['prob_sys_empty'] = prob_sys_empty
df['system_utilization'] = system_utilization
def Prob_n_C (n):
    Prob_nc = (df.system_utilization)**n * df.prob_sys_empty
    return Prob_nc

for i in range(1,30):
    df[f'n_equal_{i}'] = Prob_n_C(i)
df["arrival Rate"] = num_arrival_rate

df_new = df.set_index("arrival Rate")

df_max = df_new[0:15]
df_max.index
```

```python
plt.figure(figsize= (11,10))
for i in range(1,30):
    plt.plot(df_max[f'n_equal_{i}'], label = f'Prob cust n= {i}')
#plt.axvline(29,label = 'Maximum Capacity', color = 'k') 

plt.title("Probability of at least n customers in the system")
plt.xlabel("Arrival rates")
plt.ylabel("Probability of at least n customers")
plt.legend()
plt.savefig("different_n.png")
plt.savefig("ncustomers.png")
plt.show()
```
